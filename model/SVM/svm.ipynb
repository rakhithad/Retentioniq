```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.inspection import permutation_importance
import pickle
import joblib
```

```python
# Read your datasets
X_train = joblib.load("X_train.pkl")
X_test = joblib.load("X_test.pkl")
y_train = joblib.load("y_train.pkl")
y_test = joblib.load("y_test.pkl")
```

```python
# Creating an SVM classifier with the best hyperparameters
svm_classifier = SVC(C=10, degree=3, gamma='scale', kernel='rbf')
```

```python
# Fitting the model with training data
svm_classifier.fit(X_train, y_train)
```

```python
# Making predictions on the test set
y_pred = svm_classifier.predict(X_test)
```

```python
# Evaluate the model's performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Attrition', 'Attrition'],
            yticklabels=['No Attrition', 'Attrition'])
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

```python
# Permutation importance
result = permutation_importance(svm_classifier, X_test, y_test, n_repeats=10, random_state=42)
```

```python
# Plot feature importance
features = X_train.columns
importance = result.importances_mean  # Mean importance over all repeats

plt.figure(figsize=(10, 8))
top_n = 10
sorted_idx = importance.argsort()[-top_n:]
plt.barh(features[sorted_idx], importance[sorted_idx])
plt.xlabel('Feature Importance')
plt.title('Feature Importance for SVM (No Hyperparameter Tuning) - Permutation Importance')
plt.show()
```

```python
# Save the model as a .pkl file
with open("svm_model.pkl", "wb") as f:
    pickle.dump(svm_classifier, f)
```
